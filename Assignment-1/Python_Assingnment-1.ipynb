Assignment 1: Temperature Check
Problem: Write a program that takes the current temperature as input and prints:
● "It's too cold!" if below 10°C
● "It's cool outside" if between 10°C and 25°C
● "It's hot outside!" if above 25°C



```python
input_temp = int(input("Enter temperature: "))

if input_temp < 10:
    print("It's too cold!")
elif input_temp <= 25:
    print("It's cool outside")
else:
    print("It's hot outside!")
```

# Assignment 2: ATM Withdrawal System
# Problem: Ask the user for their account balance and the amount they want to withdraw.
 ● If the withdrawal amount is greater than the balance → print "Insufficient Balance"
 ● If the withdrawal amount is less than or equal to balance → print "Transaction Successful" and show the remaining balance.


```python
balance = 1000

print('''
      \n********** Welcome to ATM Withdrawal System **********\n
      1. Balance Check
      2. Withdraw
      3. Exit
      ''')
choice = input("Enter your choice (1-3): ")

if choice == '1':
    print(f"\nYour current balance is : {balance}\n")
    
elif choice == '2':
    withdraw_amount = int(input("\nEnter your withdraw amount: "))
    if withdraw_amount > balance:
        print("\nInsufficient Balance!\n")
    else:
        print("\nTransaction Successful\n")
        remain_balance = balance - withdraw_amount
        print(f"Now your remaining balance is : {remain_balance}\n")
        
elif choice == '3':
    print("\nThank you for using ATM!\n")

else:
    print("\nYou must choose number from (0-3)\n")
```

# Assignment 3: Student Grade Calculator
# Problem: Input marks (0–100) and print grade based on:
 ● 90–100 → A+
 ● 80–89 → A
 ● 70–79 → B
 ● 60–69 → C
 ● Below 60 → Fail

```python
marks = int(input("Enter your marks from (0-100): "))

if 90 <= marks <= 100:
    print("A+")
elif 80 <= marks <= 89:
    print("A")
elif 70 <= marks <= 79:
    print("B")
elif 60 <= marks <= 60:
    print("C")
elif 0 <= marks <= 59:
    print("Fail")
else:
    print("You must enter the marks: (0-100) ")
```

# Assignment 4: Bus Ticket Discount
# Problem: Ask the user for age and print ticket price:
 ● Below 5 → Free
 ● 5–18 → 50% Discount
 ● 60 or above → 30% Discount
 ● Others → Full Price

```python
age = int(input("Enter your age: "))
fare = 500

if age < 5:
    print("Ticket price for the age of below 5 is : Free")
elif 5 <= age <= 18:
    print(f"Ticket price for the age of {age} is : {fare - (fare * 50/100)}")
elif 60 <= age:
    print(f"Ticket price for the age of {age} is : {fare - (fare * 30/100)}")
else:
    print(f"Ticket price for the age of {age} is : {fare}")
```

# Assignment 5: Restaurant Bill with Discount
#### Problem: Take total bill amount from the user:
 ● If the bill is above 1000 → apply 10% discount
 ● Otherwise → no discount
 Then print the final bill amount.

```python
input_bill = input("Enter bill amount: ")
input_bill = float(input_bill)

discount = 10 * (1 / 100)

if 1000 < input_bill:
    discount = (input_bill * discount)
    payable = input_bill - discount 
    print(f"The final bill amount : {payable}")
else:
    print("No discount")
```

# Assignment 8: Leap Year Checker
# Problem: Take a year as input and check:
 ● If it is divisible by 400 → Leap Year
 ● Else if divisible by 4 but not by 100 → Leap Year
 ● Otherwise → Not a Leap Year


```python
year = int(input("Enter a year: "))

if year % 400 == 0:
    print("Leap Year")
elif (year % 4 == 0) and (year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

# Assignment 10: Electricity Bill Calculator
# Problem: Input number of units consumed and calculate bill:
 ● Up to 100 units → 5 BDT per unit
 ● 101–200 units → 7 BDT per unit
 ● Above 200 units → 10 BDT per unit
 Print total payable amount.

```python
units = int(input("Enter the number of units: "))

if units <= 100:
    bill = units * 5
elif units <= 200:
    bill = units * 7
else:
    bill = units * 10

print(f"Total payable amount: {bill}")
```
